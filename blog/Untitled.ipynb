### 从一个困惑开始聊起

我小时候生活在一个北方小镇，我初中的时候一个年级有四个班，初中的小孩正处于一个很敏感的时期，在这个时期他们亟需外界的认可，他们容易被情绪裹挟，他们也并不了解社会运转的规则。

我初中时候小孩会打群架。打群架的原因有很多，兄弟义气，争风吃醋，不一而足。我观察到一个有趣的现象，就是打群架的群体构成往往是以班级为单位打群架，我实际上很好奇，为什么三班要和四班打架。

这可能对很多人来说不是一个问题，但我印象很深的一件事是，由于我父母是教育系统里面的人，所以我知道分班是如何分配的，简而言之，就是随机分配的。我有可能被分到三班，也有可能被分到四班，我不理解，为什么这么随机的事情，就决定了我的朋友是谁，我的敌人是谁。这和我在动画片里看到的世界很不一样：在动画片里，主角做好事，所以主角是好人。反派做坏事，所以反派是坏人，好人要打坏人。但在现实生活中，我的朋友和我的敌人是随机决定的，如果我的考试成绩高一分，或者我的名字字母排序变一下，我就可能和我的敌人做朋友，我现在的朋友就成了我的敌人？这部荒谬吗？这太荒谬了。



我对这种荒诞现实的注意和反思让我逐渐很不认同这种随机性，例如，我坚定的不认为同一个宿舍就一定要做兄弟，同一个班级就一定要患难与共。这也促进我进一步考虑，我应该怎么选择我的朋友，后来我培养了自己一套选择朋友的方法论。但据我观察，似乎只有我觉得这件事很荒谬，其他人却乐此不疲。大部分人不考虑这种问题，大家喜欢用所谓的缘分美化这种命运的随机安排，然后慌忙的把自己塞进某个随机的集体里，随随便便称兄道弟。

当然，加入群体会有很多好处，即使是不分青红皂白的就加入本班级的打架队伍讨伐别的班级，也是有好处的。一个群体的人会因为一起征战而彼此认可，不管是打赢了还是打输了，都能消解所有人的孤独，也满足所有人的自我实现的需要。

但我想，这也是有坏处的，最大的坏处之一就是长期使用这种思维生活会模糊人的是非观。即你不再会通过一个人的行为或者一件事情本身来给出判断，而是会通过立场给出判断，也就是我们常说的屁股会决定脑袋。例如，我们经常会在互联网上看到这样的言论：因为一件事情是美国或者日本做的，无论结果怎样，他一定是错的或者邪恶的。我看着这些人引用的错误数据和阴谋论的观点，仔细看仔细看，发现写的原来是“我们是七年级三班的同学，而美国和日本是七年级四班的同学，所以自然我们是对的，他肯定是错的，我们只是需要论证这一点而已。”

曾经有一段时间，我们在一个班，所以大家就不会揪着彼此的小辫子不放了。

我不是要批评这种观点，我只是想要展示我的观察和想法，接下来我会试图从历史的视角分析这种现象形成的原因。

我出生，生长在中国北方，这是一篇很神奇的土地，至今，这片土地都没有发达的商品经济，政治权利和少量经济权力都被贵族垄断。而且这片土地还长期受到北方少数民族的袭扰，这种“袭扰”对王朝来说可能只是袭扰，但对人民来说就是事关生死的烧杀抢掠，对普通人来说，这是一种深刻的苦难。当你把视角带入到一个古代北方的普通人的时候，就会突然发现，上文中被我们展示的不符合现代性的生活策略，反而成了贫民能用的唯一策略。

抱团，不顾一切的抱团，抱大腿。

当几乎所有的资源都被少数人掌握的时候，你只有抱紧它，才能活下去，才能喝点汤。
