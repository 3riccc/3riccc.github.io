#### 我不喜欢自我意识

关于自我意识的研究，我一直处于淡淡的跟踪的状态。在情感上，我感觉这是一个有趣的话题，但只是“有趣”，尤其是受到集智俱乐部的影响，我一直关注着两批人的争论，一部分人认为意识是重要的，他们还发展了很多理论，比如整合信息论，全局工作空间理论等等。另一部分人完全认为自我意识就是幻觉，他们还做了很多实验来进行证实，比如裂脑人试验。

除了在情感上感觉这个问题很有趣，实际上我可能更倾向于自我意识幻觉派，我对这玩意一直提不起太大的感觉。或者说，我认为意识可能存在，但这不是一个重要的问题。例如，我一直隐隐感觉，如果我捕捉到了产生智能的结构和动力学规则，再通过合理的方式设置系统的输入输出，或许自我意识也就自然的产生了。

也有很多科学家在走其他的路，比如吴恩达，他捕捉了人类智慧的一些本质的机制，例如类比能力，然后他设计这种机制的算法实现，相比而言，这条路更加的实干，这比虚无缥缈的意识问题更加吸引我。

在机器学习领域里面，也没人认为意识是一个重要的问题。大家关注如何优化模型的结构和参数，从而让模型可以把握数据中所呈现的规律，然后泛化到新的数据上以解决问题。或许把前文中的“把握”替换成建模更合适，因为“把握”这个词会误导读者认为模型真的能神秘地知道些什么，就像人类在某个时刻悟到了一些什么一样。实际上，模型只是在目标函数的驱动下，数学地更改了某些内部的计算过程。而人类“悟到了”什么的时候，我们并不知道人类的大脑里发生了什么。

#### 目标函数的功绩和局限

但最近我的一些思想实验似乎告诉我，意识是一个重要的问题。这个思想实验起源于一些对更好的人工智能架构设计的幻想。

我是一个机器学习的研究者，并且一直力图用机器学习方法帮助解决一些复杂系统的核心问题。机器学习中有一个核心概念，那就是目标函数。实际上，我们通过给某个系统赋予合适运算架构、可调节参数和目标函数，在优化算法的帮助下让系统不断调整其内部的计算逻辑，从而最终引导系统达到一个看起来很智能的状态。这代表通过不断的优化调整，系统已经变成了一个非常适用于解决一类问题的灵动架构。

这种用目标函数引导系统调整的思路具有很大的市场，不仅是因为机器学习已经取得了巨大的成功，还因为这种目的论的观点和很多其他领域的观点可以遥相呼应。例如，在物种尺度，这个观点对应了自然选择，自然选择认为，如果一个种群能够不断存在并发展【目标函数】，那么这个种群就是优秀的。甚至在经济学中，人们叶发明了一个概念叫做效用函数【目标函数】，认为人的存在目的，就是在不断的最大化其效用函数，我们的种种行为，都直接或间接的关联了我们的效用函数。目标函数这个观点可以在本质上和多个学科的观点互相印证、具有良好的数学性质、且搭配上优化算法之后，可以帮助我们实实在在的解决很多问题。你可以想象目标函数为什么会这么受欢迎。

但我想在这里强调一下，本质上，这种观点是在设计一种解决特定问题的灵活架构而已，有时候我们关注的问题过大（例如学习语言的pattern），而机器学习的能力又过强，让我们产生了他具有智能，甚至通用智能的幻觉。而我们想要设计的智能系统，在我和机器学习玩了这么多年后，我越来越感觉似乎并不能通过目标函数的驱动而获得。

<hr/>

#### 目标函数难以引导复杂系统

为什么我认为目标函数不能引导我们设计真正的智能系统。这是因为，首先，我认为真正的智能系统必然是一个复杂系统，而非一个灵动但机械的系统。其次，根据有限的经验，我发现真正的复杂系统很难被一个目标函数所引导。

关于第一点不再赘述，关于第二点，为什么目标函数难以引导复杂系统。这是因为我所见到的几乎所有机器学习中的目标函数都是清晰的，而且是有限的。而复杂系统的目标往往是模糊的，多样的。我必须承认，关于这一点我只是在陈述自己的感受，而并没有清晰的理论支撑。但我们可以用一些例子来说明。

例如，我们可以这样粗略的认为，低级生物是简单的系统，而高级生物是复杂的系统。物竞天择适者生存就是作用在所有生物上的目标函数。把目光放在不同物种的尺度上，我们就会发现，越低级的生物，越被目标函数所限制，越高级的生物，似乎就越跳出了目标函数的限制。病毒和细菌唯一的目标似乎就只有繁衍，这是完全被自然选择驱动的直接行为；一只小狗在草地上撒欢的时候。你很难想象他的撒欢行为和自然选择有什么关系（你可能会说，撒欢促进好心情，好心情促进好身体，好身体促进吸引异性的能力，从而促进基因的传递，好吧，但这也太间接了）。最后让我们看人类，人类实在是太奇怪了，不但越发达的国家出生率越低，而且出现了一些“朝闻道，夕可死矣”的人。这可完全无法用自然选择解释了。

社会经济系统也是一样，你会发现，管理者往往可以对社会系统进行一定的引导，但这个引导是非常有限的。如果管理者试图将社会引向一个单一的目标函数，那么往往面临着社会经济的崩溃。苏联的极端计划经济是一个例子，中国也曾有类似的悲剧。以至于后面改革开放时，邓小平说“看不见的手”和“看得见的手”必须两手并用。我对这句话的理解是在一定程度上尊重社会经济系统自身的力量（不要给社会经济系统增加特定的目标函数），而非用看得见的手的政策力量对整个社会进行强制导向（导向某个具体的目标函数）。

科研系统（或者说学术共同体）往往被认为是一个演化的复杂系统，实际上你会发现学术共同体也没有特定的目标函数。这里有一个误区需要澄清：有人认为，我们是为了造核武器，才研究粒子物理。我们为了造光刻机，才研究光学。我们为了计算，才研究数学。所以我们需要定下某个目标，然后分解任务，然后科研攻关，实际上这只是追赶式的科研。回顾科学的历程，那些引领式的研究完全不是这样做出来的。无数的科学家并没有想的特别远，他们只是想知道这个现象，是为什么。他在想这个的时候，或许只是想解释世界，而他并不知道这东西有什么应用场景，能把人们推向哪个目标。事实上，在少量星光璀璨的理论背后。大量无人问津的理论也正在黯然退场，没有将人们推向任何目标。

#### 从目标函数的困境到自我意识

理解了目标函数无法引导复杂系统这一困境，让我们回过头来看，如果我们真的要制造一个智能系统。那么除了用目标函数引导他，我们还能怎么办？现在，这个问题看似令人毫无头绪，在这种时候，我们往往需要对一个具体的，典型的系统进行分析和观察，并且试图从中得到一些insight。接下来就让我们来以一个实在的智能系统的思考过程为例，来分析当一个智能系统没有被自然选择的目标函数所引导的时候，他到底在干什么？

实不相瞒，这个智能系统就是小时候的我。我在不同的阶段都思考过不同的问题，之所以选择一个小时候的案例，是因为那时候我还没有积累任何知识。可以排除知识对一个智能系统的干扰。

在我小时候，我有一个塑料兵团，每个塑料小人大概身高两厘米，大家有的举着枪，有的扛着炮。我会为他们排兵布阵，享受作为指挥官的乐趣。

有一次，我的一个塑料小人在桌子上。我不小心碰到了它，它啪的一下，很快啊，就落在了地上。我看着这个现象，感觉很诡异。因为我觉得从塑料小人的视角来看，桌子对他来说是一个巨大的，高高在上的悬崖。他从悬崖上跳下来，怎么能啪的一下，很快的就掉在悬崖底下呢？我从小具有丰富的翻墙经验，按照我的经验，从一座墙上跳下来，花费的时间都比塑料小人长得多。也就是说，对于塑料小人来说，他跳下一个高高的悬崖，迅速就到底了。而我只是从墙上跳下来，竟然要花比他更长的时间到底，这简直是太奇怪了。而且这和我看的玩具总动员之类的电影也不一样，在玩具总动员中，一个玩具跳下桌子，花的时间是很久的。这个世界对幼年得我来说，简直是太奇怪了。

最终，我得出了这样的结论。首先，玩具总动员肯定是假的，那是为了让我们人类看起来比较舒服而瞎编的。而我们真实世界的规则是真的。那么真实世界的规则是什么呢：很显然，对于很大的物体，他感觉到的时间流逝是很慢的。对于很小的物体，他感觉到的时间流逝是很快的。比如我感觉到的一秒钟，相当于塑料小人感觉到的一分钟。这样如果我跳下悬崖，可能我感觉花了一分钟。小人跳下桌子，对小人来说，他也感觉花了一分钟。很快，我的结论也在其他系统上得到了验证，我观察蚂蚁，我发现，蚂蚁的行为简直是神速，它在前进的时候，一秒走了特别多步，我都数不清。我想，一定是因为我的一秒钟等于蚂蚁的一分钟，所以他才能在我认知的一秒里前后摆动这么多次它的腿，给我一分钟，我也能。

这就是我小时候进行的一次“科研”，当然科研结论是一塌糊涂的，但这件事情看起来相当有趣。让我们来进行一下分析：首先，我可以保证我当时所做的所有思考都和生存繁衍没有关系，这只是单纯的好奇。而且现在没有任何研究结论证明一个好奇的人会再繁衍方面占有优势。这就意味着，在这个时候，我的行为并不是由自然选择这一目标函数引导的。那么这时候，引导我的是什么呢？我为什么会对这个问题产生好奇呢？实际上，我最初感到好奇的原因是，我感觉《从塑料小人视角来看，跳下桌子等于跳下悬崖》，我对跳桌子花的时间感到奇怪，是因为当我带入塑料小人的视角后，遇到的现象和我过往的经验产生了冲突，这是一切好奇的起点。

那么这个起点的发生，也就是说让我好奇，需要什么必要条件呢？第一点，需要儿时的我有一个明确的边界，我知道我是我，我不是小人。也就是说，需要我有一个自我意识。第二点，我在带入小人的视角，也就是说，在这个过程中，我在幻想如果我是小人，我会看到什么。这本质上是在用我的生活经验在对小人进行建模（建模一词值得商榷）。值得注意的是，这个过程并不是什么力量或目标驱动的，而更像是一个小朋友的后台进程。也就是说，一个小朋友并不需要在某种刺激下才会开始对周边物体进行建模，而只要他在和世界互动，他就会是自然地，不停地对各种东西进行自我带入。想象一个小朋友看到下雨的时候说“云彩哭了”，云彩怎么会哭呢？是因为小朋友会哭，而他用自己的行为去解释了云彩的行为。

一个训练有素的科学家，很少用自己去带入其他的系统了。我想，这并不是因为科学家的好奇机制改变了。而是科学家有了更多的知识储备，知道应该用一个什么样的模型去带入什么样的系统。当实验结果和科学家的预期不一样的时候，好奇仍然在发生着。例如做机器学习的时候，我们常常遇到这样的情况：首先，根据我的了解，模型具有这样的能力，应该在这个数据集上具有这样的表现。然后，我发现实际情况的表现不如预期。进而，我就产生了好奇，于是使用ABCD等等方法分析其原因。似乎，这和塑料小人掉悬崖，没有什么本质的区别。

#### 设计真正的智能机器？

这就是我为什么认为自我意识是一个重要的话题的原因了：因为经过前面的案例分析。我们发现，一个具有自我意识的模型不断在环境中套用自我意识，才是智能的关键。当然，目前为止这只是一个假设。我们应该如何设计一个toy model来检验这一假设呢？

很遗憾，我并不清楚，但我想在设计这个实验之前，我们有一些知识需要学习。例如，现在的意识科学认为，什么样的模型是有自我意识的，什么样的模型是没有的？我们设计实验时，是否要完全抛弃目标函数，还是把自我意识和某个机器学习的目标函数结合起来？为什么？如果目标不重要，那么衡量实验的指标是什么呢？问题多多，还需很多学习、交流和探索，本次阶段性的总结就到这里。
